# Matplotlib basics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
%matplotlib inline
```

```python
#making a basic line graph
temp=[40.6,43.9,48.1,55.7,60.8,68.7,75.4,69.6,65.8,60.4,50.1,45.6]
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
fig=plt.figure()
plt.plot(months,temp)
#adding title and label in a graph
plt.title('Average monthly tempratures')
plt.xlabel('Months')
plt.ylabel('Temprature')
plt.show
```

```python
#Saving the figure in different file formats which will get saved in this directory
fig.savefig('avgmonthlytemp.png')
```

```python
fig.savefig('avgmonthlytemp.pdf')
```
